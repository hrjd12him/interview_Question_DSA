````python
#Infix Expression: a+b*(c^d)
#Postfix Expression: abcd^*+

def infix_to_postfix(infix_expr):
    # Helper function to determine the precedence of operators
    def precedence(operator):
        if operator == '+' or operator == '-':
            return 1
        elif operator == '*' or operator == '/':
            return 2
        elif operator == '^':
            return 3
        else:
            return 0  # For parentheses

    result = ""
    stack = []

    for char in infix_expr:
        if char.isalnum():
            result += char  # Append operands to the result
        elif char == '(':
            stack.append(char)
        elif char == ')':
            while stack and stack[-1] != '(':
                result += stack.pop()
            stack.pop()  # Remove the opening parenthesis
        else:
            # Operator encountered
            while stack and precedence(char) <= precedence(stack[-1]):
                result += stack.pop()
            stack.append(char)

    # Pop any remaining operators from the stack
    while stack:
        result += stack.pop()

    return result

# Example usage
infix_expression = "a+b*(c^d)"
postfix_expression = infix_to_postfix(infix_expression)
print("Infix Expression:", infix_expression)
print("Postfix Expression:", postfix_expression)
````

### Write a function to implement Quicksort on Doubly LinkedList

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def partition_new(start, end):
    pivot = start.data
    i = start.next
    j = end

    while i != j:
        while i != j and i.data <= pivot:
            i = i.next
        while i != j and j.data >= pivot:
            j = j.prev
        if i != j:
            i.data, j.data = j.data, i.data

    start.data, j.data = j.data, start.data
    return j

def quicksort_dll(start, end):
    if end is not None and start != end and start != end.next:
        pivot = partition_new(start, end)
        quicksort_dll(start, pivot.prev)
        quicksort_dll(pivot.next, end)

def print_dll(head):
    while head:
        print(head.data, end=' ')
        head = head.next
    print()

# Example usage
def convert_array_to_dll(arr):
    head = Node(arr[0])
    current = head

    for i in range(1, len(arr)):
        new_node = Node(arr[i])
        current.next = new_node
        new_node.prev = current
        current = current.next

    return head, current

arr = [12, 4, 5, 6, 7, 3, 1, 15]
dll_head, dll_tail = convert_array_to_dll(arr)

print("Original Doubly Linked List:")
print_dll(dll_head)

# Sort the Doubly Linked List using Quicksort
quicksort_dll(dll_head, dll_tail)

print("\nDoubly Linked List after Quicksort:")
print_dll(dll_head)
````

The number of structurally unique binary trees for a given number of nodes can be calculated using the concept of Catalan numbers. The nth Catalan number, denoted by Cn, represents the number of structurally unique binary trees with n nodes.

The formula for calculating the nth Catalan number is given by:

\[ Cn = \dfrac{(2n)!}{(n+1)! \cdot n!} \]

Here's a Python function to find the number of structurally unique binary trees for a given number of nodes:

```python
def num_trees(n):
    if n <= 0:
        return 0

    # Initialize an array to store the number of unique BSTs for each number from 1 to n
    dp = [0] * (n + 1)
    dp[0] = 1
    dp[1] = 1

    # Use dynamic programming to fill in the array
    for i in range(2, n + 1):
        for j in range(1, i + 1):
            dp[i] += dp[j - 1] * dp[i - j]

    return dp[n]
```

# LRU cache
Certainly! The term "LRU" stands for "Least Recently Used," and it is often used in the context of caching algorithms. In the LRU caching strategy, the idea is to keep track of the order in which elements are accessed. When the cache reaches its capacity and a new element needs to be added, the least recently used element is evicted from the cache.

Here's a brief explanation of how LRU caching works:

1. **When an element is accessed or added to the cache:**
   - If the element is already in the cache, it is marked as the most recently used.
   - If the element is not in the cache, and the cache is not full, the new element is added and marked as the most recently used.
   - If the cache is full, the least recently used element is evicted, and the new element is added and marked as the most recently used.

2. **When an element is accessed, its position is updated:**
   - The accessed element is moved to the front of the usage order to mark it as the most recently used.
   - The least recently used element is at the back of the usage order.

The LRU caching strategy is commonly implemented using a combination of a doubly linked list and a hash map. The linked list keeps track of the order in which elements are used, and the hash map provides fast access to the elements.

Here's a basic outline of the LRU Cache implementation:

Implementing an LRU (Least Recently Used) Cache involves creating a data structure that stores a fixed number of items and evicts the least recently used item when the cache is full. We can use a combination of a doubly linked list and a hash map to efficiently implement an LRU Cache.

Here's a simple implementation in Python:

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.prev = None
        self.next = None

class LRUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.cache = {}
        self.head = Node(0, 0)  # Dummy head node
        self.tail = Node(0, 0)  # Dummy tail node
        self.head.next = self.tail
        self.tail.prev = self.head

    def _add_node_to_head(self, node):
        node.prev = self.head
        node.next = self.head.next
        self.head.next.prev = node
        self.head.next = node

    def _remove_node(self, node):
        prev = node.prev
        nxt = node.next
        prev.next = nxt
        nxt.prev = prev

    def _move_to_head(self, node):
        self._remove_node(node)
        self._add_node_to_head(node)

    def get(self, key):
        if key in self.cache:
            node = self.cache[key]
            self._move_to_head(node)
            return node.value
        else:
            return -1

    def put(self, key, value):
        if key in self.cache:
            node = self.cache[key]
            node.value = value
            self._move_to_head(node)
        else:
            if len(self.cache) == self.capacity:
                # Evict the least recently used node (from the tail)
                del self.cache[self.tail.prev.key]
                self._remove_node(self.tail.prev)

            # Add a new node to the head
            new_node = Node(key, value)
            self.cache[key] = new_node
            self._add_node_to_head(new_node)
```
